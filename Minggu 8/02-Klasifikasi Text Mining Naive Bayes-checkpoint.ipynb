# Preprocessing

## setting fungsi

```python
def bersih_data(text):
    return " ".join(text_processor.pre_process_doc(text))

def non_ascii(text):
    return text.encode('ascii', 'replace').decode('ascii')

def remove_space_alzami(text):
    return " ".join(text.split())

def remove_emoji_alzami(text):
    return ' '.join(re.sub("([x#][A-Za-z0-9]+)"," ", text).split())

def remove_tab(text):
    return text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")

def remove_tab2(text):
    return re.sub('\s+',' ',text)

def remove_rt(text):
    return text.replace('RT'," ")

def remove_mention(text):
    return ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())

def remove_incomplete_url(text):
    return text.replace("http://", " ").replace("https://", " ")

def remove_single_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

def change_stripe(text):
    return text.replace('-'," ")

def lower(text):
    return text.lower()

def remove_punctuation(text):
    remove = string.punctuation
    remove = remove.replace("_", "") # don't remove hyphens
    pattern = r"[{}]".format(remove) # create the pattern
    return re.sub(pattern, "", text) 
```

```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r'rt\s+', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

dataset['ProcessedText'] = dataset['Tweet'].astype(str).apply(clean_text)
```

```python
final_string = []

for text in df['Tweet'].astype(str).values:
    proc = text.lower()                              # lowercase
    proc = re.sub(r'rt\s+', '', proc)               # hapus 'rt'
    proc = re.sub(r'http\S+|www\S+', '', proc)     # hapus URL
    proc = re.sub(r'@\w+', '', proc)                # hapus mention
    proc = re.sub(r'#\w+', '', proc)                # hapus hashtag
    proc = re.sub(r'\d+', '', proc)                 # hapus angka
    proc = re.sub(r'[{}]'.format(re.escape(string.punctuation)), '', proc)  # hapus punctuation
    proc = re.sub(r'\s+', ' ', proc).strip()       # hapus spasi berlebih
    final_string.append(proc)

df['ProcessedText'] = final_string
```

```python
import pandas as pd

# Baca file Excel kamus perbaikan kata
normalized_word = pd.read_excel("kamus perbaikan kata.xlsx")

#  Buat dictionary normalisasi sesuai kolom Excel
normalized_word_dict = {row['TIDAK BAKU']: row['BAKU'] for index, row in normalized_word.iterrows()}

# Fungsi normalisasi token
def normalized_term(document):
    """
    Ganti setiap token dalam list 'document' sesuai kamus normalisasi.
    Jika token tidak ada di kamus, tetap dipertahankan.
    """
    return [normalized_word_dict[token] if token in normalized_word_dict else token for token in document]

# Terapkan normalisasi ke kolom token
df_new['tokens_perbaikan'] = df_new['tokens'].apply(normalized_term)

# Cek hasil
print(df_new[['tokens', 'tokens_perbaikan']].head())
```

```python
# Hapus baris dengan label kosong
dataset_clean = dataset.dropna(subset=['Sentimen'])  # ganti 'Sentimen' sesuai kolom label

# ambil fitur dan label
dataset_feature = dataset_clean['ProcessedText']
dataset_label   = dataset_clean['Sentimen']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

vectorizer = TfidfVectorizer()
dataset_feature_tfidf = vectorizer.fit_transform(dataset_feature)

train_x, test_x, train_label, test_label = train_test_split(
    dataset_feature_tfidf, dataset_label, test_size=0.3, random_state=42
)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

classifier_nb = MultinomialNB()
classifier_nb.fit(train_x, train_label)

y_pred_nb_train = classifier_nb.predict(train_x)
accuracy_nb_train = accuracy_score(train_label, y_pred_nb_train)
print("Accuracy Training set: ", accuracy_nb_train)

y_pred_nb_test = classifier_nb.predict(test_x)
accuracy_nb_test = accuracy_score(test_label, y_pred_nb_test)
print("Accuracy Test set: ", accuracy_nb_test)
```

```python
recall_nb_train = recall_score(train_label, y_pred_nb_train, average='weighted')
print("Recall Training set: ", recall_nb_train)

recall_nb_test = recall_score(test_label, y_pred_nb_test, average='weighted')
print("Recall Test set: ", recall_nb_test)
```

```python
from sklearn.metrics import precision_score

precision_nb_train = precision_score(train_label, y_pred_nb_train, average='weighted', zero_division=0)
precision_nb_test  = precision_score(test_label, y_pred_nb_test, average='weighted', zero_division=0)

print("Precision Training set: ", precision_nb_train)
print("Precision Test set: ", precision_nb_test)
```

```python
f1_nb_train = f1_score(train_label, y_pred_nb_train, average='weighted')
print("F1 Training set: ", f1_nb_train)

f1_nb_test = f1_score(test_label, y_pred_nb_test, average='weighted')
print("F1 Test set: ", f1_nb_test)
```
