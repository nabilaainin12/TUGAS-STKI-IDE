```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def PlotWordCloud(text):
    wc = WordCloud(
        width=800,
        height=400,
        background_color='white',
        collocations=False
    ).generate(text)

    plt.figure(figsize=(10,5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```
